```python
# Import dependencies
import pandas as pd
import datetime
import os
```

```python
# Create new DataFrame
new_df = pd.DataFrame()

# For each file in "Resources" directory, append its contents to the
# end of new_df
directory = "Resources/"
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    if os.path.isfile(f):
        df_to_add = pd.read_csv(directory + filename)
        new_df = pd.concat([new_df, df_to_add], ignore_index=True)

new_df
```

```python
# Drop all rows with null data
new_df = new_df.dropna()
new_df
```

```python
# Check data types
new_df.info()
```

```python
# Create a copy of new_df
new_df_copy = new_df.copy()
```

```python
# Create a new "months" column by extracting the month from each start date
months = []
for index, row in new_df_copy.iterrows():
    months.append(datetime.datetime.strptime(row["started_at"], "%Y-%m-%d %H:%M:%S").month)

new_df_copy["month"] = months
new_df_copy
```

```python
# Convert started_at and ended_at to datetime
new_df_copy["started_at"] = pd.to_datetime(new_df_copy["started_at"])
new_df_copy["ended_at"] = pd.to_datetime(new_df_copy["ended_at"])
new_df_copy.info()
```

```python
# Export DataFrame to CSV
new_df_copy.to_csv("2023-tripdata.csv")
```
