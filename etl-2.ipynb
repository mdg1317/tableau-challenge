```python
# Import dependencies
import pandas as pd
import datetime
```

```python
# Read CSV into DataFrame
citibike = pd.read_csv("2023-tripdata.csv")
citibike.head()
```

```python
# Empty lists to store station information
station_names = []
station_ids = []
station_lats = []
station_lons = []
jan_counts = []
feb_counts = []
mar_counts = []
apr_counts = []
may_counts = []
jun_counts = []
jul_counts = []
aug_counts = []
sep_counts = []
oct_counts = []
nov_counts = []
dec_counts = []

# Add 0s to each month count array
def add_new_counts():
    jan_counts.append(0)
    feb_counts.append(0)
    mar_counts.append(0)
    apr_counts.append(0)
    may_counts.append(0)
    jun_counts.append(0)
    jul_counts.append(0)
    aug_counts.append(0)
    sep_counts.append(0)
    oct_counts.append(0)
    nov_counts.append(0)
    dec_counts.append(0)

def increment_month(month, index):
    # Increase defined month count by 1
    match(month):
        case 1:
            jan_counts[index] += 1
        case 2:
            feb_counts[index] += 1
        case 3:
            mar_counts[index] += 1
        case 4:
            apr_counts[index] += 1
        case 5:
            may_counts[index] += 1
        case 6:
            jun_counts[index] += 1
        case 7:
            jul_counts[index] += 1
        case 8:
            aug_counts[index] += 1
        case 9:
            sep_counts[index] += 1
        case 10:
            oct_counts[index] += 1
        case 11:
            nov_counts[index] += 1
        case 12:
            dec_counts[index] += 1

# Iterate over rows of DataFrame
# If station ID does not exist in station_ids, add all of its data to appropriate arrays
# If it does, increment the specific month of the ride by 1
for index, row in citibike.iterrows():
    if row["start_station_id"] in station_ids:
        increment_month(datetime.datetime.strptime(row["started_at"], "%Y-%m-%d %H:%M:%S").month, station_ids.index(row["start_station_id"]))
    if row["end_station_id"] in station_ids:
        increment_month(datetime.datetime.strptime(row["started_at"], "%Y-%m-%d %H:%M:%S").month, station_ids.index(row["end_station_id"]))
    if not row["start_station_id"] in station_ids:
        station_ids.append(row["start_station_id"])
        station_names.append(row["start_station_name"])
        station_lats.append(row["start_lat"])
        station_lons.append(row["start_lng"])
        add_new_counts()
        increment_month(datetime.datetime.strptime(row["started_at"], "%Y-%m-%d %H:%M:%S").month, station_ids.index(row["start_station_id"]))
    if not row["end_station_id"] in station_ids:
        station_ids.append(row["end_station_id"])
        station_names.append(row["end_station_name"])
        station_lats.append(row["end_lat"])
        station_lons.append(row["end_lng"])
        add_new_counts()
        increment_month(datetime.datetime.strptime(row["started_at"], "%Y-%m-%d %H:%M:%S").month, station_ids.index(row["end_station_id"]))
```

```python
# Create new DataFrame with the station data
stations_df = pd.DataFrame(data={"station_id": station_ids,
    "station_name": station_names, "lat": station_lats, "lon": station_lons,
    "jan_count": jan_counts, "feb_count": feb_counts, "mar_count": mar_counts, "apr_count": apr_counts,
    "may_count": may_counts, "jun_count": jun_counts, "jul_count": jul_counts, "aug_count": aug_counts,
    "sep_count": sep_counts, "oct_count": oct_counts, "nov_count": nov_counts, "dec_count": dec_counts})
stations_df
```

```python
# Export DataFrame to CSV
stations_df.to_csv("stations.csv")
```
